# Average (weighted) Selling Price (ASP) per seller

```python
#filtro de precio
df_filtrado = df[(df['price'] <= 1000000) & (df['sold_quantity'] >0)]
df_filtrado.head()
```

```python
total_sold_qty_per_seller = df_filtrado.groupby('meli_seller_id')['sold_quantity'].sum().reset_index().rename(columns={'sold_quantity':'total_sold_qty'})
total_sold_qty_per_seller.head()
```

```python
df_filtrado = df_filtrado.merge(total_sold_qty_per_seller, on = 'meli_seller_id')
df_filtrado['ponderacion_sold_qty'] = df_filtrado.sold_quantity/df_filtrado.total_sold_qty
df_filtrado['ponderacion_precio'  ] = df_filtrado.price*df_filtrado.ponderacion_sold_qty
df_filtrado.head()
```

```python
ASP = df_filtrado[['meli_seller_id', 'total_sold_qty', 'ponderacion_precio']].groupby(['meli_seller_id', 'total_sold_qty'])['ponderacion_precio'].sum().reset_index().rename(columns={'ponderacion_precio':'ASP'})
ASP['ASP'] = ASP['ASP'].round(1)
ASP.head()
```

```python
#nombre de los sellers

sellers_name = pd.read_excel('G:\My Drive\Proyectos\[CL Scrape] Full Meli Scrape\V0\\all_meli_sellers_data.xlsx', usecols=['meli_seller_id', 'meli_seller_name', 'sales_last60days'])
sellers_name[sellers_name['meli_seller_id'] == 258161576]
```

```python
ASP_final = pd.DataFrame(ASP.merge(sellers_name, on='meli_seller_id').sort_values(by='sales_last60days',ascending = False)).set_index('meli_seller_id').fillna(0)
ASP_final.to_excel("AverageSellingPrice_WeightedBySoldQty.xlsx")
```

```python
df2 = df.merge(meli_categories, how='left', left_on=['category_id'], right_on=['id'])[['meli_item_id', 'meli_seller_id','sold_quantity', 'Shopee ID']]
df2['Shopee ID'] = df2['Shopee ID'].astype('int')
df2.head()
```

```python
shp_cat_id_to_coalesce = pd.read_excel('G:\My Drive\Proyectos\[CL Scrape] Full Meli Scrape\V0\shp_cat_id_to_coalesce (just shp_id scraped).xlsx', usecols=['global_be_category_id', 'coalesce'])
shp_cat_id_to_coalesce.head()
```

```python
df3 = df2.merge(shp_cat_id_to_coalesce, how='left', left_on=['Shopee ID'], right_on=['global_be_category_id'])[['meli_item_id', 'sold_quantity', 'meli_seller_id','coalesce']]
df3.head()
```

```python
coalesce_to_shp_cluster = pd.read_excel('G:\My Drive\Proyectos\[CL Scrape] Full Meli Scrape\V0\coalesce_to_shp_cluster.xlsx')
coalesce_to_shp_cluster.head()
```

```python
df4 = df3.merge(coalesce_to_shp_cluster, how='left', left_on=['coalesce'], right_on=['coalesced_id'])[['meli_item_id', 'meli_seller_id', 'sold_quantity', 'bi_cluster']]
df4.head()
```

```python
df_grouped_by_meli_seller_id = pd.pivot_table(data=df4[df4['sold_quantity']>0],
                                        index  =['meli_seller_id'],
                                        columns=['bi_cluster'], 
                                        values =['sold_quantity'],
                                        aggfunc=['count'],
                                        fill_value=np.nan
                                    )
df_grouped_by_meli_seller_id.to_excel('count_soldQty_per_cluster_just_sellers_who_have_sales.xlsx')
```
